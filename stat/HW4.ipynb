## Task 1

$X_1, X_2, \ldots, X_n \sim$ i.i.d. $U[0, \theta]$

$H_0: \theta = \theta_0 \\ H_1: \theta < \theta_0$

$S(X_1, X_2, \ldots, X_n) = \{X_{(n)} \leq c\theta_0\}$

Определим $c_0$:

$P_{\theta \in \Theta_0}(X \in S) = P_{\theta_0}(X \in S) = P_{\theta_0}(X_{(n)} \leq c\theta_0) = \frac{(c\theta_0)^n}{\theta_0^n} = c^n = \alpha$

Заметим, что из определения критерия, есть ограничение на $c \geq \frac{x_{(n)}}{\theta_0}$, иначе критерий никогда не будет выполняться. Также заметим, что $\theta_0 > X_{(n)}$, иначе альтернативная гипотеза никогда бы не была верна(при ее предположении, вероятность получить данную выборку была бы 0). Отсюда еще ограничение на $c$: $c \leq 1$. Иначе критерий выполняется всегда.

Таким образом, делаем вывод, что и на размер выборки есть некоторые ограничения. Пусть нам задали уровень значимости $\alpha$. Из ограничений на $c$ делаем вывод, что должно быть верно $\alpha = c^n \geq (\frac{x_{(n)}}{\theta_0})^n \to \ln{\alpha} \geq n \ln{\frac{x_{(n)}}{\theta_0}} \to n \geq \frac{\ln{\alpha}}{ \ln{x_{(n)}} - \ln{\theta_0}}$ ($\ln{\frac{x_{(n)}}{\theta_0}}$ < 0). То есть существует зависимость $\alpha$ и $n$, и при заданном уровне значимости должна быть выборка соответствующего размера.

При выполнении всех условий, можно сделать вывод о $c_0$: $c_0 = \sqrt[n]{\alpha}$.

Получается, р.н.м.к. уровня $\alpha$ будет $S(X_1, X_2, \ldots, X_n) = \{X_{(n)} \leq \sqrt[n]{\alpha}\theta_0\}$

Если рассмотреть множество $\Theta_0$ - оно состоит из 1 элемента. А вот множество $\Theta_1$ из контиинума. $\Theta_1 \in [X_{(n)}, \theta_0)$. Соответсвенно, если взять некоторое $\theta_1 \in \Theta_1$, то мы получим $f_1(X) = \frac{1}{(\theta_1)^n}$ - функция распределения для множества $X$. Использовав М.М.П., легко найти максимум - $\frac{1}{X_{(n)}^n}$ в точке $X_{(n)}$. Таким образом, полученный критерий $S(X_1, X_2, \ldots, X_n) = \{X_{(n)} \leq c\theta_0\}$ легко выводится из критерия отношений правдоподобий. Поэтому это р.н.м.к.

## Task 2

$X_1, X_2, \ldots, X_n \sim$ i.i.d. $\exp\{\theta\}$

$H_0: \theta = \theta_0 \\ H_1: \theta > \theta_0$

Экспоненциальное распределение имеет вид $f(x) = \theta e^{-\theta x}$. Пусть $f_0(X), f_1(X)$ - плотности распределения при предположении, что верна гипотеза $H_0$ и $H_1$ соответственно.

$f_1(X) = \prod\limits_{i=1}^n f_1(x_i) = (\theta_1)^n e^{-\theta_1 \sum x_i}$, где $\theta_1 > \theta_0$ - параметр из области альтернативной гипотезы.

$f_0(X) = \prod\limits_{i=1}^n f_0(x_i) = (\theta_0)^n e^{-\theta_0 \sum x_i}$

Воспользуемся критерием Неймана-Пирсона:

$\{\frac{f_1(X)}{f_0(X)} \geq \lambda \} = \{\frac{(\theta_1)^n e^{-\theta_1 \sum x_i}}{(\theta_0)^n e^{-\theta_0 \sum x_i}} \geq \lambda \} = 
\{ e^{(\theta_0 - \theta_1)\sum x_i} \geq \lambda (\frac{\theta_0}{\theta_1})^n \} = \{(\theta_0 - \theta_1)\sum x_i \geq \ln(\lambda (\frac{\theta_0}{\theta_1})^n)\} = \{\sum\limits_{i=1}^n x_i \leq \frac{\ln(\lambda (\frac{\theta_0}{\theta_1})^n)}{(\theta_0 - \theta_1)}\} = \{T_n(X) \leq t\}$

Тут $\sum\limits_{i=1}^n x_i = T_n(X)$. Это р.н.м.к., так как этот критерий общий для всех $\theta \in [\theta_0, \infty]$. Ну а также потому, что экспоненциальное распределение из экспоненциального семейства, и их критерий имеет как раз такой вид.

Чтобы найти критерий в явном виде, надо найти соответсвующий t: $P_{\theta_0}(T_n(X) \leq t) = \alpha$. Как известно, сумма i.i.d. экспоненциальных распределений имеет гамма распределение. В нашем случае $T_n(X) \sim \Gamma(n, \frac{1}{\theta_0})$. Таким образом, уравнение превращается в $F(t) =\alpha \to t = F^{-1}(\alpha)$, где $F$ - функция распределения $\Gamma(n, \frac{1}{\theta_0})$. Получается t - квантиль уровня $\alpha$ этого распределения.

Таким образом, в этом случае р.н.м.к. будет $S = \{\sum\limits_{i=1}^n x_i \leq t\}$, где $t$ - квантиль уровня $\alpha$ распределения $\Gamma(n, \frac{1}{\theta_0})$.

Посчитаем эмпирические частоты: $p_0 = \frac{36}{252}, p_1 = \frac{53}{252}, p_2 = \frac{35}{252}, p_3 = \frac{26}{252}, p_4 = \frac{30}{252}, p_5 = \frac{44}{252}, p_6 = \frac{28}{252}$. Наша гипотеза - что вероятность попасть в больницу не зависит от дня недели. Сформулируем гипотезы.

$H_0: \forall i: p_i = \frac{1}{7} \\ H_1: \exists i: p_i \neq \frac{1}{7}$

Посчитаем статистику Пирсона: $\mathcal{X}_n^2 = n \sum\limits_{i=1}^m \frac{(p_i - p_i^0)^2}{p_i^0}$, где $p_i^0=\frac{1}{7}$

$n \sum\limits_{i=1}^m \frac{(p_i - p_i^0)^2}{p_i^0} = 252 * 7 * \frac{1}{252^2}*((36 - 36)^2 + (36 - 53)^2 +(36 - 35)^2 +(36 - 26)^2 +(36 - 30)^2 +(36 - 44)^2 +(36 - 28)^2) = \\ = \frac{1}{36} * (0+289+1+100+36+64+64) = \frac{554}{36} = 15\frac{7}{18}$

Нам дан уровень значимости $\alpha = 0.05$. Степеней свободы у нас $m-1=7-1=6$. По таблице значений получаем, что $\mathcal{X}_{0.05}(6) = 12.592$.

Получаем $\mathcal{X}_n^2> \mathcal{X}_{0.05}(6)$, т.к. $15\frac{7}{18} > 12.592$. Таким образом, мы отвергаем гипотезу $H_0$ о равенстве вероятностей.

## Task 4

$X \sim B(2, \theta)$

Провели $n = 128$ испытаний. Частоты успехов: $0: \mu_1 = \frac{n}{4}$, $1: \mu_2 = \frac{n}{4}$, $2: \mu_3 = \frac{n}{2}$.

Найдем $\theta$ из о.м.п.:

$L(x, \theta) = (C_2^0\theta^0(1-\theta)^2)^{\mu_1} * (C_2^1\theta^1(1-\theta)^1)^{\mu_2} * (C_2^2\theta^2(1-\theta)^0)^{\mu_3} = 2^{\mu_2} * \theta^{\mu_2 + 2\mu_3} * (1-\theta)^{2\mu_1+\mu_2}$

$ln(L(x, \theta)) = \mu_2 ln(2) + (\mu_2 + 2\mu_3)ln(\theta) + (2\mu_1+\mu_2)ln(1-\theta)$

$\frac{\partial L(x, \theta)}{\partial \theta} = \frac{\mu_2 + 2\mu_3}{\theta} - \frac{2\mu_1+\mu_2}{1-\theta} = 0 \to \theta^{*} = \frac{\mu_2+2\mu_3}{2(\mu_1+\mu_2+\mu_3)} = \frac{5}{8}$

Таким образом, посчитаем оценочные вероятности: $0: p_1 = (1-\theta^{*})^2 = \frac{9}{64}, 1: p_2 = C_2^1\theta^{*}(1-\theta^{*}) = 2 * \frac{5}{8} * \frac{3}{8} = \frac{15}{32}, 2: (\theta^{*})^2 = \frac{25}{64}$.

Перейдем к критерию Пирсона:

$\mathcal{X}_n^2 = 128*(\frac{(0.25 - \frac{9}{64})^2}{\frac{9}{64}} + \frac{(0.25 - \frac{15}{32})^2}{\frac{15}{32}} + \frac{(0.5-\frac{25}{64})^2}{\frac{25}{64}}) = 27.876$

В нашем случае $r=1, m-r-1=3-1-1=1$. Уровень значимости $\alpha$. Если $\mathcal{X}_n^2 < \mathcal{X}_{\alpha}(1)$, то гипотеза принимается. Но в нашем случае значение статистики так велико, что оно больше почти всех квантилей распределения хи квадрат, даже при уровне значимости 0.01. 

Так что отвергаем гипотезу, что $X \sim B(2, \theta)$.

## Task 5

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./_data_.csv')
control = data[data['group'] == 'control']['data'].values
test = data[data['group'] == 'test']['data'].values
n = len(control)
```

```python
sns.histplot(data=data, x='data', hue='group')

plt.show()
```

```python
diffs = []
m = 10**4

for _ in range(m):
    control_bootstraped = np.random.choice(control, size=n, replace=True)
    test_bootstraped = np.random.choice(test, size=n, replace=True)

    control_mean = np.mean(control_bootstraped)
    test_mean = np.mean(test_bootstraped)

    diffs.append(control_mean - test_mean)
```

```python
alpha = 0.05

diffs_sorted = sorted(diffs)
left_ind = int(m * alpha / 2)
right_ind = int(m * (1 - alpha / 2))
left_point = diffs_sorted[left_ind]
right_point = diffs_sorted[right_ind]
```

```python
fig, axes = plt.subplots(1, 2, figsize =(15, 5))

graph1 = sns.histplot(diffs, ax=axes[0])
graph1.axvline(left_point, color='r', ls='--', label='левый конец д.и.')
graph1.axvline(right_point, color='black', ls='--', label='правый конец д.и.')
graph1.legend()

graph2 = sns.lineplot(x=diffs_sorted[left_ind:right_ind], y=np.zeros(right_ind - left_ind))
graph2.axvline(left_point, color='r', ls='--', label='левый конец д.и.')
graph2.axvline(right_point, color='black', ls='--', label='правый конец д.и.')
graph2.axvline(0, color='blue', ls='--', label='ноль')
graph2.legend()

plt.show()
```

Как видно по графикам, точка 0 находится далеко за пределами полученных разностей, и особенно за пределами доверительного интервала. Таким образом делаем вывод о различии средних тестовой и контрольной групп.

### В папке utils написал вспомогательные файлы.

1) category_to_num - способы перевести категориальные фичи в числовые, чтобы запустить юмап и PCA
2) custom_loss - реализация таргетной метрики
3) local_manipulations - файл с семинара, чтобы анализировать бустинг
4) plot_tool - графики для фичей
5) prepare_data - генерация фичей и подготовка данных
6) presort_data - отсортировать данные по дате
7) remove_duplicates - в датасете были дубликаты сессии. Как правило, сессия пересекала некоторую временную метку в 7-8 часов. Наверно, это новый день в Америке.
8) create_sint_train\test - файлы чтобы посчитать фичи для подхода из 1 места. VS code ест много сил, поэтому быстрее посчитать файлы.
9) генерация трейн\валидации для моего подхода

Порядок:

1) train_test_dist_research - здесь смотрю на распределение фичей трейна и теста
2) feature_research1 - здесь смотрю на респределение таргета относительно фичей.
3) research cut - здесь я привожу мой подход к решению проблемы, а также анализирую предсказуемость покупки. Здесь секция ПОДХОДЫ К ЗАДАЧЕ.
4) research_regression_cut - ПАЙПЛАЙН. И финальная модель предсказания.
5) agg_attempts - реализация топ 1 из соревы
